scipy.stats.t.sf (abs (x), df)  - основная формула

где:

x: t-оценка
df: Степени свободы

```python
import scipy.stats

#find p-value
scipy.stats.t.sf(abs(-.77), df=45)  #  p-значение, связанное с t-оценкой -0,77 и df = 45
```

```python
import scipy.stats

#find p-value
scipy.stats.t.sf(abs(1.87), df=15)# p-значение, связанное с t-показателем 1,87 и df = 15 в правостороннем тесте гипотезы
```

```python
import scipy.stats

#find p-value for two-tailed test
scipy.stats.t.sf(abs(1.24), df=22)*2
 #p-значение, связанное с t-показателем 1,24 и df = 22 в двустороннем тесте гипотезы
```
